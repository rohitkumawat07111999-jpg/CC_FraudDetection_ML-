```python
# let's see if wee need all this information and bigger float/integer types, we can improve it I believe
def optimize_numeric_columns(df):
    for col in df.columns:
        if df[col].dtype == "int64":
            df[col] = pd.to_numeric(df[col], downcast="integer")
        elif df[col].dtype == "float64":
            df[col] = pd.to_numeric(df[col], downcast="float")
optimize_numeric_columns(df)
```

```python
# okay running the pandas profiling above didn't help too much
# Split the data into features and target
# Let's use sklearn's mutual information to get some information about features
from sklearn.feature_selection import mutual_info_classif

X = df.drop("Class", axis=1)
y = df["Class"]

mi = mutual_info_classif(X, y)

N = 20
for i in np.argsort(mi)[::-1][:N]:
    print(X.columns[i], mi[i])
    
selected_features = np.argsort(mi)[::-1][:N]
```

```python
X.columns[selected_features] # these are our features that carries the most information about our target
```

```python
X.shape
```

```python
from imblearn.over_sampling import SMOTE 
# as was described in dataset description, our dataset is unbalanced and might require a balancing.

X, y = SMOTE().fit_resample(X, y)

X.shape
```

```python
y.shape
```

```python
# our features are floats and a sclaing should help us about that.
from sklearn.preprocessing import StandardScaler

# Create a scaler object
scaler = StandardScaler()

# Fit the scaler to the training data
scaler.fit(X)
```

```python
# Transform the training and test data
X = scaler.transform(X)
```
